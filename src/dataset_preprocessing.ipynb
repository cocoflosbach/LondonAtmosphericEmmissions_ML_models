```python
# Extract data from other_pollutant file into a dataframe
other_pollutants_df = pd.read_excel('LAEI2013_Emissions_Summary_OtherPollutants_v1.0.xlsb', engine='pyxlsb', sheet_name=['SO2', 'NMVOC', '1,3 Buta', 'NH3', 'C6H6', 'CO', 'CH4', 'N2O', 'Cd', 'Hg', 'Pb', 'BaP', 'PCB', 'HCl'], header=3, usecols=['Total'])
```

```python
# Extract all other pollutangts into dataframes
#SO2
SO2_df= other_pollutants_df.get('SO2')
SO2_df.rename(columns= {'Total': 'SO2'}, inplace = True)

NMVOC_df= other_pollutants_df.get('NMVOC')
NMVOC_df.rename(columns= {'Total': 'NMVOC'}, inplace = True)

Buta_df= other_pollutants_df.get('1,3 Buta')
Buta_df.rename(columns= {'Total': '1,3 Buta'}, inplace = True)

NH3_df= other_pollutants_df.get('NH3')
NH3_df.rename(columns= {'Total': 'NH3'}, inplace = True)

C6H6_df= other_pollutants_df.get('C6H6')
C6H6_df.rename(columns= {'Total': 'C6H6'}, inplace = True)

CO_df= other_pollutants_df.get('CO')
CO_df.rename(columns= {'Total': 'CO'}, inplace = True)

CH4_df= other_pollutants_df.get('CH4')
CH4_df.rename(columns= {'Total': 'CH4'}, inplace = True)

N2O_df= other_pollutants_df.get('N2O')
N2O_df.rename(columns= {'Total': 'N2O'}, inplace = True)

Cd_df= other_pollutants_df.get('Cd')
Cd_df.rename(columns= {'Total': 'Cd'}, inplace = True)

Hg_df= other_pollutants_df.get('Hg')
Hg_df.rename(columns= {'Total': 'Hg'}, inplace = True)

Pb_df= other_pollutants_df.get('Pb')
Pb_df.rename(columns= {'Total': 'Pb'}, inplace = True)

BaP_df= other_pollutants_df.get('BaP')
BaP_df.rename(columns= {'Total': 'BaP'}, inplace = True)

PCB_df= other_pollutants_df.get('PCB')
PCB_df.rename(columns= {'Total': 'PCB'}, inplace = True)

HCl_df= other_pollutants_df.get('HCl')
HCl_df.rename(columns= {'Total': 'HCl'}, inplace = True)
```

```python
Cd_df
```

```python
#Solution 2
all_pollutant_dfs= [PM10_df, PM2_5_df, NOx_df, CO2_df, SO2_df, NMVOC_df, Buta_df, NH3_df, C6H6_df, CO_df, CH4_df, N2O_df, Cd_df, Hg_df, Pb_df, BaP_df, PCB_df, HCl_df]
merged_pollutant_df = pd.concat(all_pollutant_dfs, join='inner', axis=1)
```
